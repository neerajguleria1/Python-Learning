```python
# Cell 4 — write driver logs to a file
drivers = [
    {"name": "Neeraj", "speed": 180, "seatbelt": True},
    {"name": "Amit",  "speed": 220, "seatbelt": False},
    {"name": "Sujal", "speed": 140, "seatbelt": True},
]

with open("drivers_log.txt", "w", encoding="utf-8") as f:
    for d in drivers:
        if d["speed"] > 200 and not d["seatbelt"]:
            f.write(f"{d['name']} - 🚨 Overspeeding & no seatbelt\n")
        elif d["speed"] > 150:
            f.write(f"{d['name']} - ⚠️ Overspeeding\n")
        else:
            f.write(f"{d['name']} - ✅ Driving safely\n")

print("drivers_log.txt created.")
!ls -la driver_log.txt
```
