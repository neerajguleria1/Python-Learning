```python
# Cell 4 â€” write driver logs to a file
drivers = [
    {"name": "Neeraj", "speed": 180, "seatbelt": True},
    {"name": "Amit",  "speed": 220, "seatbelt": False},
    {"name": "Sujal", "speed": 140, "seatbelt": True},
]

with open("driver_log.txt","w",encoding="utf-8") as f:
  for d in drivers:
    if d["speed"]>200 and not d["seatbelt"]:
      f.write(f"{d['name']}- Overspeeding and no seatbelt ")
    elif d["speed"]>150:
      f.write(f"{d['name']}-Overspeeding ")
    else:
      f.write(f"{d['name']} is going normally ")

print("driver_log.txt created successfully")
!ls -la driver_log.txt
```

```python
new_event={"name":"Arpit","speed":330,"seatbelt":True}

with open("driver_log.txt","a",encoding="utf-8") as f:
  if new_event["speed"]>220 and not new_event["seatbelt"]:
    f.write(f"{new_event['name']} - Overspeeding ")
  else:
    f.write(f"{new_event['name']}- going normally")
print("appended new_event to driver_log.txt")

with open("driver_log.txt","r",encoding="utf-8") as f:
  print(f.read())
```
