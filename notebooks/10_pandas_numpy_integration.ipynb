```python
# Data Cleaning

import numpy as np
import pandas as pd

data={
    'Driver':['Neeraj','Chandra','Nishant','Amit','Arpit',np.nan],
    'Speed':[190,130,np.nan,445,249,550],
    'Mileage':[18,np.nan,27,8,14,'seven'],
    'Alertness':['Medium','Low','Medium',np.nan,'Low','High']
}
df=pd.DataFrame(data)
print("Data is:\n",df)

df['Driver']=df['Driver'].fillna('Unknown')
df['Speed']=df['Speed'].fillna(df['Speed'].mean())
df['Mileage']=pd.to_numeric(df['Mileage'],errors='coerce')
df['Mileage']=df['Mileage'].fillna(df['Mileage'].mean())
df['Alertness']=df['Alertness'].fillna('Medium')

df=df.drop_duplicates()
print("Cleaned data:\n",df)
print(df.dtypes)

df['Effeciency']=df['Mileage']/(df['Speed']/100)
print(df)
df_sorted=df.sort_values(by='Effeciency',ascending=False)
hgst_Eff=df_sorted.head(1)
print(hgst_Eff)
df.to_csv("Cleaned_driver_data.csv")
```

```python
# Day17
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("Cleaned_driver_data.csv")
print("Data preview\n",df.head())
plt.figure(figsize=(8,5))
plt.bar(df['Driver'],df['Speed'],color='skyblue')
plt.title("Driver Speeds")
plt.xlabel("Driver")
plt.ylabel("Speed (km/h)")
plt.show()

plt.figure(figsize=(8,5))
plt.bar(df['Driver'],df['Mileage'],color='orange')
plt.title("Driver Mileage")
plt.xlabel("Driver")
plt.ylabel("Mileage (km/h)")
plt.show()

plt.figure(figsize=(7,5))
plt.hist(df['Speed'],bins=5,color='green',alpha=0.7)
plt.title("Speed Distribution")
plt.xlabel("Speed")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,5))
plt.plot(df['Driver'],df['Effeciency'],marker='o',linestyle='-',color='red')
plt.title("Driver Effeciency trend")
plt.xlabel("Driver")
plt.ylabel("Effeciency")
plt.show()
```
