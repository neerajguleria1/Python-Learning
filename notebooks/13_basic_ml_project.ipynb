```python
#Reverse String
class Solution:
  def reverseString(self,s:str)->str:
    s=list(s)
    left,right=0,len(s)-1
    while left<right:
      s[left],s[right]=s[right],s[left]
      left+=1
      right-=1
    return ''.join(s)

print(Solution().reverseString("Neeraj"))
```

```python
#Driver Above Average Speed
import pandas as pd
import numpy as np

df=pd.DataFrame({
    'Driver':['Neeraj','Sujal','Arpit','Nishant','Amit'],
    'Speed':[130,66,86,170,200]
})
avd_speed=np.mean(df['Speed'])
fast_driver=df[df['Speed']>avd_speed]

print("Average Driver:",avd_speed)
print("\n Fastest Driver:\n",fast_driver)
```

```python
# Closest Speeds
import numpy as np
speeds=np.array([120,408,55,26,662])
diff_matrix=np.abs(speeds.reshape(-1,1)-speeds).astype(float)
np.fill_diagonal(diff_matrix,np.inf)
closest_pair=np.unravel_index(np.argmin(diff_matrix),diff_matrix.shape)
print("Closest pair indices",closest_pair)
```

```python
# Fill Missing values
import pandas as pd
import numpy as np

df=pd.DataFrame({
    'Speed':[120,np.nan,200,249,np.nan],
    'Mileage':[13,55,np.nan,23,61]
})
df['Speed']=df['Speed'].fillna(df['Speed'].mean())
df['Mileage']=df['Mileage'].fillna(df['Mileage'].mean())
df=df.astype(int)
print(df)
```

```python
import numpy as np

speeds=df['Speed']
mean=np.mean(speeds)
std=np.std(speeds)
outliers=speeds[speeds>mean+2*std]
print(outliers.tolist())
```
