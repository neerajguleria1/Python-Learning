```python
# Task
''' Driver fuel effeciency Tracker (Using Pandas)'''

import numpy as np
import pandas as pd

mileage=np.array([
    [10,12,50,23],
    [17,12,66,21],
    [13,22,16,22],
    [50,72,58,28],
    [90,52,5,26]
])
drivers=["Neeraj","Chandra","Arpit","Nishant","Sujal"]
df=pd.DataFrame(mileage,index=drivers,columns=['Trip1','Trip2','Trip3','Trip4'])

df['AvgMileage']=df.mean(axis=1)
#print(df)

df['Category']=np.where(df['AvgMileage']>25,'Efficient','Needs Improvement')
df_sorted=df.sort_values(by='AvgMileage',ascending=False)

top2=df_sorted.head(2)
top2.to_csv('top2_driver.csv')
print("Full Driver Data:\n",df_sorted)
print("Top 2 effecient drivers are",top2)
```
